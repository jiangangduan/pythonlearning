```python
import transformers
import warnings
warnings.filterwarnings("ignore")
```

```python
#we refer to code here  https://www.machinecurve.com/index.php/2021/02/16/easy-machine-translation-with-machine-learning-and-huggingface-transformers/
#and the model link is https://huggingface.co/Helsinki-NLP/opus-mt-zh-en

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Initialize the tokenizer
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-zh-en")

# Initialize the model
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-zh-en")

def do_translate(text):
  # Tokenize text
  # text = "我慢慢来到拱形校门下，踏着遍地落叶，看着迤逦的阳光拖成长长的身影，缓缓回首，千种情思涌上心头。"
  tokenized_text = tokenizer.prepare_seq2seq_batch([text], return_tensors='pt')

  # Perform translation and decode the output
  translation = model.generate(**tokenized_text)
  translated_text = tokenizer.batch_decode(translation, skip_special_tokens=True)[0]

  return translated_text
```
